#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here
# Pivot function is opposite to melt.
# The syntax is 
# in the index argument you put the columns you want to keep
# in the columns argument you put the columns whose values you want to make into columns
dataset_pivot = employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = {'Salary':'mean'})
dataset_pivot.head()
```

```python
dataset_groupby = employee.groupby(['Department']).agg({'Salary':np.mean})
dataset_groupby
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2
# your answer here
dataset_pivot = employee.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = {'Salary':'mean'})
dataset_pivot.head()
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
dataset_pivot = employee.pivot_table(index= ['Department','Title'], values = ['Salary','Name'], aggfunc = {'Salary':'mean','Name':'count'})
dataset_pivot.head()
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
dataset_pivot = employee.pivot_table(index= ['Department','Title'],values = ['Salary'], aggfunc = ['max','min','mean','std']).head(5)
dataset_pivot
```
